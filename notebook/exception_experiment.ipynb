```python
import sys
import traceback
```

```python
from logger.custom_logger import CustomLogger
```

__file__ in .py automatically get the name but not in ipynb

```python
logger=CustomLogger().get_logger("exception_experiment")
```

```python
class DocumentPortalException(Exception):
    #custom exception for the document portal
    def __init__(self,error_message,error_details:sys):
        _,_,exc_tb=error_details.exc_info()
        self.file_name=exc_tb.tb_frame.f_code.co_filename
        self.lineno=exc_tb.tb_lineno
        self.error_message=str(error_message)
        self.traceback_str= ''.join(traceback.format_exception(*error_details.exc_info()))        
        pass
    def __str__(self):
        return f"""
        Error occurred in file: {self.file_name} at line: [{self.lineno}] with 
        message: {self.error_message} and
        traceback: {self.traceback_str}
        """
if __name__ == "__main__":
    try:
        #simulate an error
        a = 1 / 0
        print(a)
    except Exception as e:
        app_exc=DocumentPortalException(e,sys)
        logger.error(app_exc)
        raise app_exc 
```

1. log this info (file, cloud config, command prompt) it will keep the record of error and will not stop the execution of the program.
2. raise the exception (stop the execution)
